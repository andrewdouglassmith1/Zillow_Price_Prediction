```python
test_links = home_links[21:24]
test_links
```

```python
# Create a list of failed links to circle back to and a list to feed dataframes into for all properties

failed_links_rd1 = []
list_df_rd1 = []
count = 0
```

```python
for url in home_links:
    count += 1
    print(count)
    try:
        # Load webpagfe into driver and soup
        driver = webdriver.Chrome(chromedriver)
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, "html5lib")
        
        # Initiate dictionary (we will add address information from the link format)
        property_info = {}
        property_info['price'] = {}
        property_info['bedrooms'] = {}
        property_info['bathrooms'] = {}
        property_info['sf'] = {}
        property_info['link'] = url
        property_info['year_built'] = {}
        property_info['lot_size'] = {}
        property_info['view_description'] = {}
        property_info['on_waterfront'] = {}
        property_info['house_style'] = {}
        property_info['new_construction'] = {}
        property_info['garage_spaces'] = {}
        
        # Locating the number of bathrooms
        property_info['bathrooms'] = soup.find_all("span",class_ = "ds-bed-bath-living-area")[1].text
        
        # Locating year built
        property_info['year_built'] = soup.find(class_="ds-home-fact-list").find_all("span")[3].text
        
        # Locating Lot Size
        property_info['lot_size'] = soup.find(class_="ds-home-fact-list").find_all("span")[11].text
        
        # Garage Spaces
        for element in soup.find_all(class_="Text-c11n-8-18-0__aiai24-0 foiYRz"):
            try:
                if "Garage spaces:" in element.text:
                    string = element.text
                    split_string = string.split(":", 1)
                    property_info['garage_spaces'] = split_string[1]
                    break
            except:
                pass
        
        # Locating square footage
        for i,element in enumerate(soup.find_all("h3")):
            try:
                if i == 1:
                    string = element.text
                    split_string = string.split("a", 1)
                    property_info['sf'] = split_string[1]
                    break
            except:
                pass
        
        # Locating if on waterfront
        for element in soup.find_all(class_="Text-c11n-8-18-0__aiai24-0 foiYRz"):
            try:
                if "On waterfront:" in element.text:
                    string = element.text
                    split_string = string.split(":", 1)
                    property_info['on_waterfront'] = split_string[1]
                    break
            except:
                pass
        
        # View Description
        for element in soup.find_all(class_="Text-c11n-8-18-0__aiai24-0 foiYRz"):
            try:
                if "View description:" in element.text:
                    string = element.text
                    split_string = string.split(":", 1)
                    property_info['view_description'] = split_string[1]
                    break
            except:
                pass
            
        # Architectural style
        for element in soup.find_all(class_="Text-c11n-8-18-0__aiai24-0 foiYRz"):
            try:
                if "Architectural style:" in element.text:
                    string = element.text
                    split_string = string.split(":", 1)
                    property_info['house_style'] = split_string[1]
                    break
            except:
                pass
        
        # New Construction
        for element in soup.find_all(class_="Text-c11n-8-18-0__aiai24-0 foiYRz"):
            try:
                if "New construction:" in element.text:
                    string = element.text
                    split_string = string.split(":", 1)
                    property_info['new_construction'] = split_string[1]
                    break
            except:
                pass
            
        # Bedrooms
        for element in soup.find_all(class_="Text-c11n-8-18-0__aiai24-0 foiYRz"):
            try:
                if "Bedrooms:" in element.text:
                    string = element.text
                    split_string = string.split(":", 1)
                    property_info['bedrooms'] = split_string[1]
                    break
            except:
                pass
        # Price

        for element in soup.find_all("h4"):
            try:
                if "$" in element.text:
                    string = element.text
                    property_info['price'] = string
                    break
            except:
                pass

        # Convert dictionary to pandas dataframe and save in list
        df = pd.DataFrame.from_dict(property_info,orient = 'index')
        df1 = df.T
        list_df_rd1.append(df1)
        property_info = {}
        print("The current link {} passed".format(url))
    # If fails, append link to failed links
    except:
        failed_links_rd1.append(i)
        print("The current link {} failed".format(url))
        pass
        
    # Space requests
    time.sleep(15+2*random.random())
```
